The following table summarizes some of the common notation used in
[Luke Harmon's book on phylogenetic comparative methods](https://lukejharmon.github.io/pcm).
As in any other language, context matters in mathematics;
The info in the table does not guarantee that the same symbol is not
used to represent something else somewhere in the book.


<table>
    <tr>
        <th>Notation</th>
        <th>Description</th>
    </tr>
    <tr>
        <td>
            $t$
        </td>
        <td>
            Time.
        </td>
    </tr>
    <tr>
        <td>
            $\sigma^2$
        </td>
        <td>
            Evolutionary rate parameter.
            <b>Note</b>: In statistics, $\sigma^2$ is used to represent the population variance, and throughout the book, it is used to represent various types variance.
        </td>
    </tr>
    <tr>
        <td>
            $\bar{z}(t)$
        </td>
        <td>
            The mean value of the trait of the population at time $t$
        </td>
    </tr>
    <tr>
        <td>
            $E[X]$
        </td>
        <td>
            The expected value (or mean) of the random variable $X$ ($X$ could be anything).
        </td>
    </tr>
    <tr>
        <td>
            $\sim$
        </td>
        <td>
            Distributed as. For example, $x \sim \textrm{uniform}(0, 1)$ means that $x$ is a random variable that follows a uniform distribution.
        </td>
    </tr>
    <tr>
        <td>
            $N(x, y)$
        </td>
        <td>
            A normal distribution with mean = $x$ and variance = $y$.
        </td>
    </tr>
    <tr>
        <td>
            $C$
        </td>
        <td>
            Phylogenetic variance-covariance matrix. It is a matrix representation of a phylogenetic tree.
            Each row and column represents a species on the phylogeny.
            Each entry along the diagonal is the root-to-tip distance for a species.
            The off-diagonal entries are the total length of branches shared by the species being compared.
        </td>
    </tr>
    <tr>
        <td>
            $R$
        </td>
        <td>
            Variance-covariance matrix across traits.
            Each row and column represents a trait measured for the species being studied.
            Each entry along the diagonal is the evolutionary rate for a trait ($\sigma_{i}^{2}$).
            Each off-diagonal entry is the evolutionary covariance between the traits being compared (i.e., their tendency to evolve together).
        </td>
    </tr>
    <tr>
        <td>
            $V$
        </td>
        <td>
            Variance-covariance matrix across all traits and all species.
            Each row and column represents one of the traits for one of the species being studied.
            Each entry along the diagonal is the evolutionary rate for a trait ($\sigma_{i}^{2}$) multplied by the root-to-tip distance for a species (the expected amount of evolutionary change that happened for this trait for this species, starting from the root of the tree).
            Each off-diagonal entry is the evolutionary covariance between the traits and species being compared.
        </td>
    </tr>
</table>
